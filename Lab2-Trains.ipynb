9. As an exercise only store the trains whose traincode starts with a D

```python
for objTrainPositionsNode in objTrainPositionsNodes:
    traincodenode = objTrainPositionsNode.getElementsByTagName("TrainCode").item(0)
    if traincodenode is not None:
        traincode = traincodenode.firstChild.nodeValue.strip()
        if traincode.startswith("D"):
            dataList.append(traincode)

with open("Start_D_trains.csv", "w", newline="") as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(["TrainCode"])  # Write header row
    for code in dataList:
        csv_writer.writerow([code])

# Open and read the CSV file
with open("Start_D_trains.csv", "r") as csvfile:
    csv_reader = csv.reader(csvfile)
    
    # Read and print each row
    for row in csv_reader:
        print(row)
```
